
# 📊 Comparativo Final de Modelos de Regresión

A lo largo de este trabajo se implementaron y analizaron cuatro modelos de regresión distintos para predecir la variable **`TravelledDistance`**, basada en variables como `MatchAssists`, `MatchHeadshots`, `PrimaryHeavy`, y `RoundFlankKills`.

---

## 1. 📈 Regresión Lineal Simple
- **Variable predictora**: `MatchAssists`
- **R²**: Bajo (~0.28)
- **Interpretación**: Por cada asistencia adicional, se estiman en promedio 0.522 muertes más.
- **Ventaja**: Simplicidad y fácil interpretación.
- **Limitación**: Bajo poder explicativo y alta varianza residual.

---

## 2. 📊 Regresión Lineal Múltiple
- **Variables predictoras**: `MatchAssists`, `MatchHeadshots`, `PrimaryHeavy`, `RoundFlankKills`
- **R²**: Mejor que el modelo simple, aunque menor al de Random Forest.
- **Importancia**: Las asistencias y headshots fueron las variables con mayor peso.
- **Ventaja**: Capacidad de capturar relaciones lineales múltiples.
- **Limitación**: No captura relaciones no lineales o interacciones complejas.

---

## 3. 🌳 Árbol de Decisión
- **Modelo basado en reglas jerárquicas** que divide el espacio de decisiones según las variables predictoras.
- **R²**: Aceptable, pero algo inferior al Random Forest.
- **Ventaja**: Explicable visualmente (estructura de decisiones).
- **Limitación**: Tiende a sobreajustar si no se controla la profundidad y nodos mínimos.

---

## 4. 🌲 Random Forest
- **Modelo más robusto y preciso del análisis.**
- **R²**: Alto (>0.85), explica más del 85% de la variabilidad en `TravelledDistance`.
- **Variables más importantes**: `MatchAssists` y `MatchHeadshots`.
- **Ventajas**:
  - Excelente precisión.
  - Generaliza bien nuevos datos.
  - No necesita normalización.
  - Inmunidad al sobreajuste al combinar múltiples árboles.
- **Limitación**: Difícil interpretación directa (modelo de caja negra).

---

## 🏆 Conclusión Final

Entre todos los modelos probados, **Random Forest** demostró el mejor desempeño general en términos de precisión, robustez y capacidad de interpretación mediante importancia de características.

Los modelos lineales son útiles para relaciones simples y explicaciones directas. El árbol de decisión aporta interpretabilidad pero pierde precisión.

**Recomendación**: Para aplicaciones reales que requieren mayor exactitud, se sugiere usar Random Forest como modelo principal.
